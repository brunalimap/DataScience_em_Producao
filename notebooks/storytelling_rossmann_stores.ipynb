## <font color=#6A2626> Preparação dos Dados </font>

Esta etapa é importante para que os algortimos de Machine Learning tem um melhor aprendizados se as variáveis estão na mesma escala e se são dados numéricos. E a preparação dos dados pode ser dividida em 3 etapas sendo elas:

- **Normalização:** Tem como objetivo trazer as variáveis para o mesmo range.
    
    - Rescala o centro para 0 com desvio padrão igual a 1.
    - Verificamos quais variáveis tem distribuição normal.
    


- **Rescaling:** Nesta etapa você rescala as variáveis numéricas.

    - Rescala para o intervalo entre 0 e 1.
    - É necessário verificar se as variáveis apresenta outliers.
    

- **Transformação:** Dividido em duas etapas sendo elas a conversão das features categóricas para numéricas e a transformações de variáveis.

    - Conversão de Features Categóricas para Numéricas (Encoding).
    - Transformação de Grandeza tem como objetivo trazer a sua variável resposta o mais próximo de uma normal.
    - Tranformação de Natureza tem como objetivo trazer a natureza real dentro do seu conjunto de dados

Ao encontrar outliers foi usado o <b>RobustScaler</b> por ter um melhor desempenho para essas variáveis pois ela usa os quartis em vez do máximo e mínimo. Já para as variáveis que não possuem outliers vamos usar o <b>MinMaxScaler</b>.

```python
rs = RobustScaler()
mms = MinMaxScaler()

#competition_distance (Robust)

df5['competition_distance'] = rs.fit_transform(df5[['competition_distance']].values)

#competition_time_month (Robust)

df5['competition_time_month'] = rs.fit_transform(df5[['competition_time_month']].values)

#year (Rescale)

df5['year'] = mms.fit_transform(df5[['year']].values)

#promo_since_week (Rescale)

df5['promo_since_week'] = mms.fit_transform(df5[['promo_since_week']].values)
```

### <font color=#6A2626> Encoding </font> 

```python
# state_holiday - One Hot Encoding

df5 = pd.get_dummies(df5,prefix=['state_holiday'], columns=['state_holiday'])

# store_type Label Encoding

le = LabelEncoder()
df5['store_type'] = le.fit_transform(df5['store_type'])

# assortment Ordinal Encoding

assortment_dict = {'basic':1,'extra':2,'extended':3}
df5['assortment'] = df5['assortment'].map(assortment_dict)
```

### <font color=#6A2626> Transformação da Variável Resposta </font>  

```python
df5['sales']= np.log1p(df5['sales'])
```

### <font color=#6A2626> Transformação Natural </font>  

```python
# Variáveis ciclicas (que sempre se repetem em um determinado tempo)

#month
df5['month_sin']= df5['month'].apply(lambda x: np.sin(x * (2. * np.pi/12) ) )
df5['month_cos']= df5['month'].apply(lambda x: np.cos(x * (2. * np.pi/12) ) )

#day
df5['day_sin']= df5['day'].apply(lambda x: np.sin(x * (2. * np.pi/30) ) )
df5['day_cos']= df5['day'].apply(lambda x: np.cos(x * (2. * np.pi/30) ) )

#week_of_year
df5['week_of_year_sin']= df5['week_of_year'].apply(lambda x: np.sin(x * (2. * np.pi/52) ) )
df5['week_of_year_cos']= df5['week_of_year'].apply(lambda x: np.cos(x * (2. * np.pi/52) ) )

#day_of_week
df5['day_of_week_sin']= df5['day_of_week'].apply(lambda x: np.sin(x * (2. * np.pi/7)) )
df5['day_of_week_cos']= df5['day_of_week'].apply(lambda x: np.cos(x * (2. * np.pi/7)) )
```

## <font color=#6A2626>Métricas Usadas </font>

<b>MAE</b>
- Ele atribui peso igual a todos os erros.
- Robusto na presença de Outliers.
- Fácil entendimento pela equipe de negócios.

<b>MAPE</b>
- Mostra a que distância a previsão está do valor real, em média, como uma porcentagem.
- Amplamente utilizado para relatar resultados.
- Não pode ser usado se a variável de resposta contiver zero. Se você tiver que prever zero.

<b>RMSE</b>
- Dá muito peso aos grandes erros.
- Sensível na presença de outliers.
- Ideal para medir o desempenho do modelo de aprendizado de máquina.
